# MNIST Digit Classification with PyTorch 

##  Concept Overview

You’re training a **fully connected feedforward neural network** to classify **handwritten digits (0–9)** from the **MNIST dataset** using PyTorch.

The major steps are:

1. **Data Preparation**

   * Load dataset → Transform images into tensors → Normalize pixel values.
2. **Model Definition**

   * Create a neural network with layers and activation functions.
3. **Loss Function & Optimizer**

   * Loss: Measures how wrong the model is.
   * Optimizer: Updates model weights to reduce loss.
4. **Training Loop** (forward pass → loss → backward pass → weight update).
5. **Evaluation** (test accuracy).





`````python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
`````

`````python
# Transformations (convert to tensor + normalize)
transform = transforms.Compose([
    transforms.ToTensor(), 
    transforms.Normalize((0.5,), (0.5,))
])
`````

`````python
# Load MNIST dataset
train_dataset = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root="./data", train=False, download=True, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
`````

`````python
# Define a simple Neural Network
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(28*28, 128)  # First hidden layer
        self.fc2 = nn.Linear(128, 64)     # Second hidden layer
        self.fc3 = nn.Linear(64, 10)      # Output layer (10 digits)

    def forward(self, x):
        x = x.view(-1, 28*28)  # Flatten image
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

model = SimpleNN()
`````

`````python
# Loss & Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
`````

`````python
# Training loop
for epoch in range(5):  # Train for 5 epochs
    running_loss = 0.0
    for images, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    
    print(f"Epoch [{epoch+1}/5], Loss: {running_loss/len(train_loader):.4f}")
`````

`````python
# Evaluation
correct, total = 0, 0
with torch.no_grad():
    for images, labels in test_loader:
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Accuracy on test set: {100 * correct / total:.2f}%")
`````

`````python
from PIL import Image
import matplotlib.pyplot as plt
`````

`````python
# Load image, convert to grayscale, resize to 28x28
img = Image.open("digit.png").convert("L").resize((28, 28))
`````

`````python
# Convert to tensor and normalize
img_tensor = transforms.ToTensor()(img)
img_tensor = transforms.Normalize((0.5,), (0.5,))(img_tensor)
`````

`````python
# Predict
with torch.no_grad():
    output = model(img_tensor.unsqueeze(0))
    _, predicted = torch.max(output, 1)

print(f"Predicted Digit: {predicted.item()}")
`````

`````python
# Show the image
plt.imshow(img, cmap="gray")
plt.title(f"Predicted: {predicted.item()}")
plt.axis("off")
plt.show()
`````

This project builds and trains a simple neural network to recognize handwritten digits (0–9) from the MNIST dataset using PyTorch.

---

## 1. Imports
We bring in the required modules:
- **torch, torch.nn, torch.optim** → Core PyTorch libraries for tensor computation, defining neural networks, and optimization.

- **torch.utils.data.DataLoader** → Handles batching, shuffling, and parallel loading of data.

```python
transform = transforms.Compose([
    transforms.ToTensor(),
````

* **`ToTensor()`** → Converts image to PyTorch tensor and scales pixel values to `[0, 1]`.

---


* **`shuffle=True`** → Randomizes training data order.
